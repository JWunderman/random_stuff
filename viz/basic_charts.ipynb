# Heatmap

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.dates as mdates
import plotly.express as px


```

## Seaborn

```python
# Creating a sample DataFrame
np.random.seed(0)
data = np.random.rand(10, 12)
df = pd.DataFrame(data, columns=[f'Var{i+1}' for i in range(12)], index=[f'Obs{i+1}' for i in range(10)])

# Creating a custom colormap
cmap = ListedColormap(sns.color_palette("coolwarm", 256))

# Setting up the matplotlib figure
plt.figure(figsize=(12, 9))

# Drawing the heatmap with the mask and correct aspect ratio
sns.set(style="white")
heatmap = sns.heatmap(df, cmap=cmap, annot=True, fmt=".2f", linewidths=1, cbar_kws={"shrink": .8}, vmin=0, vmax=1)

# Adding a color bar at the top
cbar = heatmap.collections[0].colorbar
cbar.set_label('Scale', labelpad=15, fontsize=12)
cbar.ax.tick_params(labelsize=10)

# Adding title and labels
plt.title('Heatmap Chart', fontsize=20, pad=20)
plt.xlabel('Variables', fontsize=15)
plt.ylabel('Observations', fontsize=15)

# Adjusting the layout
plt.tight_layout()

# Displaying the plot
plt.show()
```

### Plotly

```python

# Creating a sample DataFrame
np.random.seed(42)
data = np.random.rand(100, 10)
df = pd.DataFrame(data, columns=[f'Var{i+1}' for i in range(10)])

# Calculating the correlation matrix
corr = df.corr()

# Creating the heatmap
fig = px.imshow(corr, color_continuous_scale='darkmint', text_auto=True, zmin=-1, zmax=1)

# Adding title
fig.update_layout(title_text='Correlation Heatmap', title_x=0.5)

# Displaying the plot
fig.show()
```

# Line Chart

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Setting the style
sns.set(style="whitegrid")

# Creating a date range
dates = pd.date_range("2023-01-01", periods=100)

# Generating three time series of random data
data_series_1 = np.random.rand(100)
data_series_2 = np.random.rand(100)
data_series_3 = np.random.rand(100)

# Creating a DataFrame
df = pd.DataFrame({
    'Date': dates,
    'Series 1': data_series_1.cumsum(),
    'Series 2': data_series_2.cumsum(),
    'Series 3': data_series_3.cumsum()
})

# Melting the data to long format
df_melted = pd.melt(df, id_vars='Date', var_name='Series', value_name='Value')

# Creating the cumulative time series line chart
plt.figure(figsize=(10, 6))
line_chart = sns.lineplot(data=df_melted, x='Date', y='Value', hue='Series', palette=['blue', 'green', 'red'], linewidth=2.5)

# Adding title and labels
plt.title('Cumulative Time Series Line Chart', fontsize=20)
plt.xlabel('Date', fontsize=15)
plt.ylabel('Cumulative Sum', fontsize=15)

# Formatting the x-axis to show months
line_chart.xaxis.set_major_locator(mdates.MonthLocator())
line_chart.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Adding a legend
plt.legend(title='Series')

# Displaying the plot
plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objects as go

# Creating a date range
dates = pd.date_range("2023-01-01", periods=100)

# Generating three time series of random data
data_series_1 = np.random.rand(100).cumsum()
data_series_2 = np.random.rand(100).cumsum()
data_series_3 = np.random.rand(100).cumsum()

# Creating a Plotly figure
fig = go.Figure()

# Adding series to the figure
fig.add_trace(go.Scatter(x=dates, y=data_series_1, mode='lines', name='Series 1'))
fig.add_trace(go.Scatter(x=dates, y=data_series_2, mode='lines', name='Series 2'))
fig.add_trace(go.Scatter(x=dates, y=data_series_3, mode='lines', name='Series 3'))

# Updating the layout of the plot
fig.update_layout(
    title='Three Timeseries Line Chart',
    xaxis_title='Date',
    yaxis_title='Cumulative Sum',
    template="plotly_white"
)

# Showing the plot
fig.show()
```

## Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Creating some synthetic data
np.random.seed(42)
x = np.random.rand(100)
y = 2 * x + 1 + 0.1 * np.random.randn(100)

# Setting the aesthetic style of the plots
sns.set_style("whitegrid")

# Creating the scatter plot
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(x, y, color="skyblue", s=100, edgecolor="w", linewidth=0.5)

# Highlighting some points
highlight = (x > 0.2) & (x < 0.8)
scatter_plot.scatter(x[highlight], y[highlight], color='red', s=100, edgecolor='w', linewidth=0.5)

# Adding titles and labels
plt.title('Cool Looking Scatter Plot', fontsize=20)
plt.xlabel('X-axis', fontsize=15)
plt.ylabel('Y-axis', fontsize=15)

# Customizing the axes
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Adding a text annotation
plt.text(0.8, 2.5, 'Highlighted Points', fontsize=12, color='red')

# Adding a grid
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

# Showing the plot
plt.show()
```

### plotly

```python
import plotly.express as px
import numpy as np

# Setting a random seed for reproducibility
np.random.seed(42)

# Creating some synthetic data
x = np.linspace(0, 10, 100)
y = 2 * x + 1 + np.random.randn(100) * 2

# Creating a scatter plot
fig = px.scatter(x=x, y=y, title='Cool Scatter Plot with Plotly',
                 labels={'x': 'X-axis', 'y': 'Y-axis'},
                 template='plotly_dark', width=700, height=500)

# Adding a trendline
fig.add_scatter(x=x, y=2 * x + 1, mode='lines', name='Trendline')

# Customizing the layout
fig.update_layout(
    font=dict(family="Arial, sans-serif", size=12, color='white'),
    title_font=dict(size=24),
    title_x=0.5,  # Centering the title
    xaxis=dict(showgrid=True, gridwidth=1, gridcolor='lightgrey'),
    yaxis=dict(showgrid=True, gridwidth=1, gridcolor='lightgrey'),
)

# Adding annotations
fig.add_annotation(
    x=8, y=17,
    text="Data Points",
    showarrow=True,
    arrowhead=1,
    ax=20,
    ay=-30
)
fig.add_annotation(
    x=2, y=5,
    text="Trendline",
    showarrow=True,
    arrowhead=1,
    ax=-20,
    ay=-30
)

# Showing the plot
fig.show()
```
